```python
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.layers import MaxPooling2D,Dense,Flatten,Conv2D
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import to_categorical
import tkinter as tk
from tkinter import Canvas
from PIL import Image,ImageDraw
from tensorflow.keras.models import load_model
```

```python
(X_train , y_train),(X_test,y_test)=mnist.load_data()
```

```python
X_train=X_train.reshape(-1,28,28,1)
X_test=X_test.reshape(-1,28,28,1)

input_shape=(28,28,1)

X_train=X_train.astype('float32') / 255
X_test=X_test.astype('float32') / 255

y_train=to_categorical(y_train,10)
y_test=to_categorical(y_test,10)
```

```python
model=Sequential()

model.add(Conv2D(32 , (3,3) , activation='relu' , input_shape=input_shape))
model.add(MaxPooling2D(2,2))
model.add(Conv2D(64 , (3,3) , activation='relu'))
model.add(MaxPooling2D(2,2))
model.add(Flatten())
model.add(Dense(128 , activation='relu'))
model.add(Dense(10 , activation='softmax'))
```

```python
model.compile(optimizer='adam' , loss='categorical_crossentropy' , metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=20 , batch_size=42 , validation_data=(X_test,y_test))
```

```python
model.save("Brain.h5")
```

```python
model = load_model('Brain.h5')

class DigitApp:
    
    def __init__(self,win):
        
        self.win=win
        self.win.config(bg="Purple")
        self.win.title("mnist")
        self.win.geometry("400x600")
        
        self.canvas=Canvas(win,width=280,height=280,bg='white')
        self.canvas.pack(pady=20)
        
        tk.Button(win,text="clear" , command=self.clear, font=("arial" , 16) , bg="#05A3CA").pack()
        tk.Button(win,text="predict" , command=self.predict, font=("arial" , 16) , bg="#05A3CA").pack()
        tk.Button(win,text="save image" , command=self.save_image, font=("arial" , 16) , bg="#05A3CA").pack()
        
        self.result=tk.Label(win,text="number : ?", font=('arial' , 20), bg="green")
        self.result.pack(pady=10)
        
        self.image=Image.new("L",(280,280),0)
        self.draw=ImageDraw.Draw(self.image)
        self.canvas.bind("<B1-Motion>" ,self.paint)
        
    def paint(self,event):
        
        x1,y1=(event.x - 10),(event.y - 10)
        x2,y2=(event.x - 10),(event.y + 10)
        
        self.canvas.create_oval(x1,y1,x2,y2, fill='black' , width=20)
        self.draw.ellipse([x1,y1,x2,y2] , fill=255)
        
    def save_image(self):
        
        self.filename = "my_drawing.jpg"
        self.image.save(self.filename)
    
    
    def clear(self):
        
        self.canvas.delete("all")
        self.image=Image.new("L",(280,280),0)
        self.draw=ImageDraw.Draw(self.image)
        self.result.config(text="number : ?")
        
        
    def predict(self):
        
        img=self.image.resize((28,28))
        img_arr=np.array(img).astype('float32') / 255
        img_arr=img_arr.reshape(-1,28,28,1)
        
        pred=model.predict(img_arr)
        app=np.argmax(pred)
        conf=np.max(pred) * 100
        
        self.result.config(text=f"number : {app}\nTrust : {conf:.1f}%")
        
        
win=tk.Tk()
app=DigitApp(win)
win.mainloop()
```
